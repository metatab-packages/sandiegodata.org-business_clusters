```python
df['cluster_type'] = df['cluster_type'].fillna('NA')
t = df.groupby(['naics_6_desc','naics_6','cluster_type']).account.count().unstack()

j = t.divide(t.sum(axis=1), axis=0).multiply(100).round(0)
j.columns = ['NA_pct','casual_pct','ent_pct','shop_pct']
t = pd.concat([t,j], axis=1)
c = 'NA'
t = t[t[c] > 10].sort_values(c+'_pct', ascending=False)
t.head(50)
```

```python
na_naics = t[t.NA_pct>60].reset_index().naics_6
na_df = df[df.naics_6.isin(na_naics)]
```
