```python
import seaborn as sns
import metapack as mp
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display 

%matplotlib inline
sns.set_context('notebook')
mp.jupyter.init()
```

```python
pkg = mp.jupyter.open_package()
#pkg = mp.jupyter.open_source_package()
pkg
```

```python
pkg.resource('sd_business_clusters')
```

```python
scb = pkg.resource('sd_custered_businesses').geoframe()
scb.head().T
```

```python
scb['popden'] = (scb['pop']  / scb['area']) * 1000**2
scb.popden.describe()
```

```python
naics = pkg.resource('naics').dataframe()
naics.head()
```

```python
scb.cluster_type.value_counts()
```

```python
df = scb.merge(naics)
import geopandas as gpd
df['geometry'] = gpd.points_from_xy(df.lon, df.lat)
df = gpd.GeoDataFrame(df, crs=4326)

# There are a lot of duplicates, mostly because a business can be in multiple clusters
df = df.drop_duplicates(subset=['dba_name','geometry'])
df['cluster_type'] = df.cluster_type.fillna('NA')
len(df)
```

```python
df.cluster_type.value_counts()
```

```python
df['pop_quintile'] = pd.qcut(df.popden, 5).cat.codes
```

```python
t = df[df.cluster_type == 'NA']
counts = t.groupby('naics_4').account.count()

common_naics = counts.sort_values(ascending=False).iloc[:50]
dfc = t[t.naics_4.isin(common_naics.index)]
t = dfc.groupby(['naics_4_desc', 'pop_quintile']).account.count().sort_values(ascending=False).unstack()
rs = t.sum(axis=1)

t = t.divide(rs,axis=0)
t = t.multiply(100).round(0).replace({np.inf:np.nan}).fillna(0).astype(int)
t.columns = 'q1 q2 q3 q4 q5'.split()
t['qr'] = t['q5']/t['q1']
t = t.sort_values('qr', ascending=False)
t
```

```python
sd_business_clusters = pkg.resource('sd_business_clusters').geoframe()
sd_business_clusters.to_file('clusters.geojson', driver='GeoJSON')
```
